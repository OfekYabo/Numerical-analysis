# Rate of convergence

```python
import numpy as np

i=np.array(range(10), dtype=np.float64)
```

### Caution!

`array` default type would be `int32`

```python
print(i)
print(i+5)
print(i*2)

l = list(range(10))
print(l)
print(l*2)
```

```python
a = 1.0/(i+1)
print(a)
b = 1.0/(2**i)
print(b)
c = 1.0/2.0**(2.0**i)
print(c)
```

```python
import matplotlib.pyplot as plt

plt.plot(i,c,"red")
plt.show()
```

```python
plt.plot(i,a,"green")
plt.plot(i,b,"blue")
plt.plot(i,c,"red")
plt.show()
```

```python
fig, (axa,axb,axc) = plt.subplots(nrows=1, ncols=3, sharex=False, figsize=(12, 4))

axa.set_title("sublinear convergence")
axa.semilogy(i,a,"green")
   
axb.set_title("linear convergence")
axb.semilogy(i,b,"blue")

axc.set_title("super linear convergence")
axc.semilogy(i,c,"red")
plt.show()
```
