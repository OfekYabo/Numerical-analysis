```python
def TIMED(function):
    from time import time
    def wrapper(*args, **kwargs):
        T = time()
        result = function(*args, **kwargs)
        T = time()-T
        print(">>>function {} timed {}".format(function.__name__,T))
        return result
    return wrapper
```

```python
@TIMED
def f(a):
    return a+1

print(f(1))
```

```python
import numpy as np

@TIMED
def sum(*args, **kwargs): 
    return sum(*args, **kwargs)

a = np.array([0.1]*10000000, dtype=np.float32)
sum(a)
```
