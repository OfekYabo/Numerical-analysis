Put the code files as described in the filesystem-tree above.
**i.e:**
1. Create a directory inside the *data_full* with your name and id seperated ny underscore (\<your name_id\>).
2. In this directory place your files of the assignments.
3. In the outer directory, place the other relevant files (grader etc.)

A small explanation:
* Every directory in *data_full* is considered a submission.
* For each submission, our testing files are copied into the submission directory.
* The *grader.py* file is then run and creates *res.csv* file with the results of your submission.

````python
import os
import subprocess

from shutil import copyfile
import pandas as pd

def main():
    dirs = set()
    # c_dir =
    c_dir=os.path.join(os.getcwd(),'data_full')

    '''
    Run testing on each directory in data_full
    (TIP: Can be used for comparison between versions, where each version is a different directory)
    '''
    dir_files=os.listdir(c_dir)
    graders=[]
    for f in dir_files:
        full_path = os.path.join(c_dir, f)
        if os.path.isdir(full_path):
            dirs.add(full_path)
    for dir in dirs:
        print(f"running on {dir}")
        copyfile(os.path.join(os.getcwd(),'grader.py'),os.path.join(dir,'grader.py'))
        copyfile(os.path.join(os.getcwd(), 'commons.py'), os.path.join(dir, 'commons.py'))
        copyfile(os.path.join(os.getcwd(), 'functionUtils.py'), os.path.join(dir, 'functionUtils.py'))
        copyfile(os.path.join(os.getcwd(), 'sampleFunctions.py'), os.path.join(dir, 'sampleFunctions.py'))
        try:
            sp=subprocess.run(['python', f'{os.path.join(dir,"grader.py")}'],timeout=1000)
        except subprocess.TimeoutExpired:
            print(f'dir {dir} has timeout')
        except Exception as e:
            print(f'dir {dir} has error {e}')


if __name__ == '__main__':
    main()
````
